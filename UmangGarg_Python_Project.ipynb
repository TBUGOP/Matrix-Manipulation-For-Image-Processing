# <font color='red'>Applying operation on matrix to flip the image</font>

```python
for i in range(len(image2)):
    image2[i].reverse()
plt.imshow(image2)
plt.show()

#Here by using a for loop we are reversing the elements in each row of matrix.
```

To achieve this task, we can simply reverse each row of each channel in the matrix of image, so reversing the row means that we are flipping the whole image. 
There are two types of flipping: UP-DOWN, RIGHT-LEFT 

For <b> Up-Down flipping we reverse the columns</b>.

For <b> Right-Left we reverse the rows</b>.

We can also flip the image along the diagonal by simply transposing the matrix of image

```python
flipped = np.fliplr(rgbMe)
plt.imshow(flipped)
plt.show()
```

###### fliplr OR flip-left-right is a numpy function which is doing the same thing that we did above. But this time, we are reversing each row in all three channels of matrix.

# <font color='red'>Multiplying various Kernels to matrix for blurring the image</font>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("IMG_1668 2.jpg")
rgbMe = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur = cv2.blur(rgbMe,(5,5)) # taking a matrix of dimension 35x35
plt.subplot(121),plt.imshow(rgbMe),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

In order to blur an image, we can just simply average out a particular pixel with its surrounding pixels. To achieve this task, we follow convolutional method where we take an imaginary matrix of any dimension it could be 3x3, 5x5, 25x25 and so on....
We name this imaginary matrix as Kernel. To blur an image, we place this kernel on each pixel of an original image and replace that pixel with an <b>average</b> of all the pixels that comes inside the kernel. <br>
<b><i>Higher the kernel dimensions mean higher will be the blur effect.</i></b>
